```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a synthetic dataset representing categories and their corresponding values
np.random.seed(42)

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))  # Random values between 10 and 100 for each category

# Create a DataFrame for better organization
df = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# Plotting the comparison using a bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Category'], df['Value'], color='lightblue', edgecolor='black')

# Add labels and title
plt.title('Comparison of Categories Based on Numerical Values', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Show plot
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generate synthetic data for categories and numerical values
np.random.seed(42)

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
category_values = []

# Generate 100 random numerical values for each category
for category in categories:
    values = np.random.normal(loc=np.random.randint(20, 50), scale=10, size=100)  # Normal distribution with random mean and std
    category_values.extend([(category, val) for val in values])

# Convert to a DataFrame
df = pd.DataFrame(category_values, columns=['Category', 'Value'])

# Visualize the distribution using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')

# Add labels and title
plt.title('Distribution of Numerical Values Across Categories', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Show plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with 5 features, where some features are correlated
n_samples = 1000

# Generate features with correlation
feature_1 = np.random.normal(loc=50, scale=10, size=n_samples)  # Normally distributed
feature_2 = feature_1 + np.random.normal(loc=0, scale=5, size=n_samples)  # Correlated with feature_1
feature_3 = feature_1 * 0.5 + np.random.normal(loc=0, scale=5, size=n_samples)  # Another feature correlated with feature_1
feature_4 = np.random.normal(loc=30, scale=10, size=n_samples)  # Independent feature
feature_5 = feature_2 * 0.6 + np.random.normal(loc=0, scale=5, size=n_samples)  # Correlated with feature_2

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4,
    'Feature 5': feature_5
})

# Compute the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, linewidths=0.5)

# Add title and labels
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(df['X'], df['Y'], df['Z'], color='b', alpha=0.7, edgecolors='k', s=40)

# Add labels and title
ax.set_title('3D Scatter Plot of Data Points', fontsize=16)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generate synthetic sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
sales_matrix = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(sales_matrix, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=0.5, cbar_kws={'label': 'Sales'})

# Add title and labels
plt.title('Variation in Sales Across Different Months and Days', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Day', fontsize=12)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate the synthetic dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.6, edgecolors="w", linewidth=2)

# Add labels to each bubble (Country name)
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, fontsize=12, ha='center', va='center')

# Add title and labels
plt.title('Bubble Chart: Population vs GDP by Country', fontsize=16)
plt.xlabel('GDP (in Billions)', fontsize=12)
plt.ylabel('Population (in Millions)', fontsize=12)

# Show plot
plt.grid(True)
plt.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate the x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate the y-values as the sine of x
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with the x and y values
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
from bokeh.io import output_file
from bokeh.models import ColumnDataSource

# Generate random x, y, sizes, and colors
np.random.seed(42)
n = 100  # Number of points
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.randint(10, 50, n)  # Random marker sizes between 10 and 50
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a ColumnDataSource from the generated data
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a new plot
p = figure(title="Random Scatter Plot with Varying Sizes and Colors", 
           x_axis_label='X', y_axis_label='Y', 
           width=800, height=400)

# Add a scatter plot to the figure using the sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```
